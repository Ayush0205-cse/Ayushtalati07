**Merging The Dataset**

```python
merge = pd.merge(deliveries,matches,left_on='match_id',right_on='id')
merge.head()
```

```python
merge.describe()
```

```python
merge.head(5)
```

**Batsmen With Highest Sixes**

```python
six = merge[merge['batsman_runs']==6]
a=six.groupby('batsman').count()['batsman_runs'].sort_values(ascending=False).head(10)
a
```

```python
plt.figure(figsize=(12,6))
sea.barplot(data=merge,x=a.index,y=a)
```

**TOP WICKET TAKERS**

```python
deliveries.groupby('bowler')['player_dismissed'].count().sort_values(ascending=False).head(10).plot(kind='bar',color='green')
plt.title('Top Wicket Taker of IPL')
plt.xlabel('Bowler')
plt.ylabel('Total Wickets Taken')
plt.show()
```
